```python
#Student Pass/Fail Classification Supervised Data using
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

data = {"StudyHours":[4,8,3,3,10,10,2,6,8,10,8,5,6], "Attendance":[10,20,40,70,90,30,90,50,70,100,80,40,50], "Assignments":[1,1,4,5,8,4,2,2,5,6,1,9,2], "Result": ["Fail","Fail","Fail","Fail","Pass","Fail","Fail","Fail","Fail","Pass","Fail","Fail","Fail"]}

df = pd.DataFrame(data)

df["Result"] = df["Result"].map({"Fail":0, "Pass":1})

X = df[["StudyHours","Attendance","Assignments"]]
y = df["Result"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, predictions))
```
